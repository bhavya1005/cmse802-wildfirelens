```python
import pandas as pd
import sqlite3

# Connect to the database
conn = sqlite3.connect("../data/raw/FPA_FOD_20170508.sqlite")

# Load main table
df = pd.read_sql("SELECT * FROM Fires", conn)

# Initial view
print("Shape:", df.shape)
df.head()
```

```python
# List all columns
df.columns.tolist()
```

```python
# Summary stats for numeric columns
df.describe()
```

```python
df = df[[
    'FIRE_YEAR',
    'DISCOVERY_DATE',
    'STAT_CAUSE_DESCR',
    'FIRE_SIZE',
    'FIRE_SIZE_CLASS',
    'LATITUDE',
    'LONGITUDE',
    'STATE'
]]
df.head()
```

```python
# Convert Julian discovery date to datetime
df['DISCOVERY_DATE'] = pd.to_datetime(df['DISCOVERY_DATE'], origin='julian', unit='D')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['FIRE_SIZE'], bins=50, log_scale=True)
plt.title("Distribution of Fire Size (Log Scale)")
plt.xlabel("Fire Size (acres)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='FIRE_SIZE_CLASS', order=sorted(df['FIRE_SIZE_CLASS'].unique()))
plt.title("Fires by Size Class")
plt.xlabel("Size Class")
plt.ylabel("Number of Fires")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='STAT_CAUSE_DESCR', order=df['STAT_CAUSE_DESCR'].value_counts().index)
plt.title("Fires by Cause")
plt.xlabel("Count")
plt.ylabel("Cause")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.sample(10000), x='LONGITUDE', y='LATITUDE', hue='FIRE_SIZE_CLASS', alpha=0.6, palette="viridis")
plt.title("Fire Locations by Size Class (sample of 10k)")
plt.show()
```

```python
# Distribution of fire sizes (log scale because it's skewed)
sns.histplot(df['FIRE_SIZE'], bins=100, log_scale=True)
plt.title("Distribution of Fire Sizes (log scale)")
plt.xlabel("Fire Size (acres)")
plt.ylabel("Number of Fires")
plt.show()
```

```python
# Count of fires by size class (A–G)
sns.countplot(data=df, x='FIRE_SIZE_CLASS', order=sorted(df['FIRE_SIZE_CLASS'].unique()))
plt.title("Fires by Size Class")
plt.xlabel("Size Class")
plt.ylabel("Count")
plt.show()
```

```python
# Causes of fire
sns.countplot(data=df, y='STAT_CAUSE_DESCR', order=df['STAT_CAUSE_DESCR'].value_counts().index)
plt.title("Fires by Cause")
plt.xlabel("Number of Fires")
plt.ylabel("Cause")
plt.show()
```

```python
# Fires per year
sns.countplot(data=df, x='FIRE_YEAR')
plt.xticks(rotation=90)
plt.title("Number of Fires per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# Quick map-like scatter plot of fires
sample_df = df.sample(10000)  # Sample for speed
sns.scatterplot(data=sample_df, x='LONGITUDE', y='LATITUDE', hue='FIRE_SIZE_CLASS', alpha=0.5, palette="viridis")
plt.title("Geospatial Distribution of Fires by Size Class (sample)")
plt.show()
```

- Most fires are in the A/B/C size class (<100 acres)
- Majority of fires caused by lightning, debris burning, campfires, and miscellaneous sources
- Fires tend to cluster in specific geolocations (e.g., California, western US)
- Very skewed fire size distribution → use log scale for modeling
- Strong temporal spread between 1992 and 2015
